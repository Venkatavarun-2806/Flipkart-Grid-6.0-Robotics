```python

import torch
from matplotlib import pyplot as plt
import cv2

# Load the YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# Path to your image
image_path = 'F:\FlaskApp\Detection\download (1).jpeg'

# Inference
results = model(image_path)

# Results
print(f"Number of objects detected: {len(results.xyxy[0])}")

# Displaying the image with bounding boxes
results.show()  # This will automatically show the image with bounding boxes

# If you'd like to use Matplotlib for better control:
# Convert results to numpy array and display with Matplotlib
img = cv2.imread(image_path)
results.render()  # Updates results with the bounding box on the image

# Convert image from BGR to RGB for Matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Plot the image
plt.imshow(img_rgb)
plt.axis('off')  # Turn off axis labels
plt.show()
```
